```python
class Node(object):
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
```

```python
def BSTSequences(root, toDo = None, soFar = None, results = None):
    if root is None:
        return
    if soFar is None:
        soFar = []
    if toDo is None:
        toDo = []
    soFar.append(root.data)
    if results is None:
        results = []
    if not toDo and not root.left and not root.right:
        results.append(soFar)
    
    if root.left:
        toDo.append (root.left)
    if root.right:
        toDo.append (root.right)
        
    for i in range (len(toDo)):
        curr = toDo.pop(0)
        BSTSequences(curr,toDo,soFar)
        toDo.append(curr)
    
    if root.left:
        toDo.pop()
    if root.right:
        toDo.pop()
        
    soFar.pop()
    
    return results
```

```python
root = Node(2)
root.left = Node(1)
root.right = Node(3)
```

```python
BSTSequences(root)
```

```python
def allSequences(root):
    results = []
    if root is None:
        results.append([])
        return results
    
    leftSeq = allSequences(root.left)
    rightSeq = allSequences(root.right)
    
    prefix = [root.data]
    
    for l in leftSeq:
        for r in rightSeq:
            weaved = []
            weave(l,r,weaved,prefix)
            results.extend(weaved)
    return results
        
        
        
def weave(first, second, results, prefix):
    
    if not first or not second:
        result = list(prefix)
        result.extend(first)
        result.extend(second)
        results.append(result)
        return
    
    save = first.pop(0)
    prefix.append(save)
    weave (first, second, results,prefix)
    prefix.pop()
    first.insert(0, save)
    
    save = second.pop(0)
    prefix.append(save)
    weave(first,second,results,prefix)
    prefix.pop()
    second.insert(0, save)
```

```python
%timeit allSequences(fifty)
%timeit BSTSequences(fifty)
```
